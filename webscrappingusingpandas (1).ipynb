```python
url = 'https://dipam.gov.in/en/disinvestments/past?field_financial_year_selection_tid=2508'
```

```python
import pandas as pd
```

```python
df = pd.read_html(url,header = 0)
df
```

# Data Cleaning

```python
df = df2017.drop(df2017[df2017.Title == 'Page SUM'].index)
df

```

```python
df1 = df.drop(df2017[df2017.Title == 'Total SUM'].index)
df1
```

```python
len(df1)
```

```python
df1.shape
```

```python
x = df1["% of GoI's Shares disinvested"].replace('-',0)
```

```python
x
```

# Quick Exploratory Data Analysis

```python
import seaborn as sns
```

### Making Histogram

```python
sns.distplot(x,kde=False)
```

### Change the bar line color

```python
sns.distplot(x,kde=False,hist_kws=dict(edgecolor="black",linewidth=2))
```

### Change the bar fill colors

```python
sns.distplot(x,kde=False,hist_kws=dict(edgecolor="black",linewidth=2),color='#00BFC4')
```
